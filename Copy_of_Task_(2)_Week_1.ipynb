## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
df.describe() # display the summary statistics of the dataframe
```

### Identify and analyze patterns in the data.

```python
df['data'] = pd.to_datetime(df['Datetime']) # convert the Datetime column to a datetime object
```

```python
print(df.groupby('City').size())
#df['City'].value_counts()
```

```python
grouped_df = df.groupby(['City']).agg({
   'JamsCount':  ['sum', 'count'],
   'JamsDelay': ['sum', 'count'],
   'JamsLength': ['sum', 'count'],
   'TravelTimeHistoric': ['sum', 'count'],
   'TravelTimeLive': ['sum', 'count']
})
print (grouped_df)
```

```python
numeric_columns = df.select_dtypes(include=['number'])
```

```python
numeric_columns.corr() # display the correlation matrix of the dataframe
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
plt.figure(figsize=(10, 5))
plt.hist(df['TrafficIndexLive'], bins=20, color='skyblue')
plt.title('Distribution of Traffic Index')
plt.xlabel('Traffic Index')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['Datetime'] = pd.to_datetime(df['Datetime'])
df['Date'] = df['Datetime'].dt.date
df['Date'].value_counts().sort_index().plot(kind='line', color='skyblue')
plt.title('Traffic Data over Time')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).

> Add blockquote




```python
plt.figure(figsize=(10, 5))
plt.scatter(df['JamsCount'], df['TrafficIndexLive'], alpha=0.5)
plt.title('Traffic Index vs Jams Count')
plt.xlabel('Jams Count')
plt.ylabel('Traffic Index')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(12, 6))
grouped_df['JamsCount']['sum'].plot(kind='bar', color='skyblue')
plt.title('Total Jams Count by City')
plt.xlabel('City')
plt.ylabel('Total Jams Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
grouped_df['JamsLength']['sum'].plot(kind='bar', color='salmon')
plt.title('Total Jams Length by City')
plt.xlabel('City')
plt.ylabel('Total Jams Length (km)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
grouped_df['JamsDelay']['sum'].plot(kind='bar', color='lightgreen')
plt.title('Total Jams Delay by City')
plt.xlabel('City')
plt.ylabel('Total Jams Delay (minutes)')
plt.xticks(rotation=45)
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.
